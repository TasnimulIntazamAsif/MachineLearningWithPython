```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data=sns.load_dataset("titanic")
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Drop rows with null values in age and deck columns
data = data.dropna(subset=['age', 'deck'])
```

```python
data.isnull().sum()
```

```python
# Drop rows with null values in embarked and embark_town columns
data = data.dropna(subset=['embarked', 'embark_town'])
```

```python
data.isnull().sum()
```

```python
data
```

```python
# Boolean encoding for categorical columns
categorical_columns = ['sex', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone']

# Create boolean encoded variables (0/1) for each categorical column
data_encoded = data.copy()

# Sex: male=1, female=0
data_encoded['sex'] = (data['sex'] == 'male').astype(int)

# Embarked: convert to numeric, S=0, C=1, Q=2
data_encoded['embarked'] = pd.Categorical(data['embarked']).codes

# Class: First=0, Second=1, Third=2
data_encoded['class'] = pd.Categorical(data['class'], categories=['First', 'Second', 'Third']).codes

# Who: child=0, woman=1, man=2
data_encoded['who'] = pd.Categorical(data['who'], categories=['child', 'woman', 'man']).codes

# Adult_male: False=0, True=1
data_encoded['adult_male'] = data['adult_male'].astype(int)

# Deck: A=0, B=1, C=2, D=3, E=4, F=5, G=6
data_encoded['deck'] = pd.Categorical(data['deck']).codes

# Embark_town: Cherbourg=0, Queenstown=1, Southampton=2
data_encoded['embark_town'] = pd.Categorical(data['embark_town']).codes

# Alive: no=0, yes=1
data_encoded['alive'] = (data['alive'] == 'yes').astype(int)

# Alone: False=0, True=1
data_encoded['alone'] = data['alone'].astype(int)

# Display the first few rows of encoded data to verify
data_encoded.head()
```
